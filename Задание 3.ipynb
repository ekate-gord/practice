```python
import xml.etree.ElementTree as ET
```

```python
def parse_ket_eaf_file(path):
    root = ET.parse(path).getroot()
    for sentence in root.findall('TIER[@TIER_ID="rus"]/ANNOTATION/REF_ANNOTATION'):
        annotation = sentence.attrib['ANNOTATION_REF']
        print("Предложение на русском: ")
        value_rus = sentence.find('ANNOTATION_VALUE')
        rus = value_rus.text
        print(rus)
        for word in root.findall('TIER[@TIER_ID="fonWord"]/ANNOTATION/REF_ANNOTATION[@ANNOTATION_REF="'+annotation+'"]'):
            annot_id = word.attrib['ANNOTATION_ID']
            print("Слововорма на кетском: ")
            value_ket = word.find('ANNOTATION_VALUE')
            ket = value_ket.text
            print(ket)
            for morph in root.findall('TIER[@TIER_ID="fon"]/ANNOTATION/REF_ANNOTATION[@ANNOTATION_REF="'+annot_id+'"]'):
                annotation_id = morph.attrib['ANNOTATION_ID']
                print("Разбор слова: ")
                value_fon = word.find('ANNOTATION_VALUE')
                fon = value_fon.text
                print(fon)
                for morph_item in root.findall('TIER[@TIER_ID="A_morph-msa-en"]/ANNOTATION/REF_ANNOTATION[@ANNOTATION_REF="'+annotation_id+'"]/ANNOTATION_VALUE'):
                    A_morph_msa_en = morph_item.text
                    print("Часть речи: ", A_morph_msa_en)
                    for item in root.findall('TIER[@TIER_ID="A_morph-type"]/ANNOTATION/REF_ANNOTATION[@ANNOTATION_REF="'+annotation_id+'"]/ANNOTATION_VALUE'):
                        A_morph_type = item.text
                        print("Stem: ", A_morph_type)
```

```python
parse_ket_eaf_file('IrikovaMM_Medvezhij prazdnik.eaf')
```
